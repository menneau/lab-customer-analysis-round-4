```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

customer = pd.read_csv('marketing_customer_analysis.csv')
```

```python
# 1 - Check the data types of the columns and separate numeric and categorical columns.

# Separate numeric and categorical columns
numerical = customer.select_dtypes(include=[np.number])
categoricals = customer.select_dtypes(include=[object])

# Display the first few rows of numerical DataFrame
print("Numerical DataFrame:")
print(numerical.head())

# Display the first few rows of categoricals DataFrame
print("\nCategoricals DataFrame:")
print(categoricals.head())
```

```python
# 2 - Visualize the distribution of numerical variables.

sns.set(style="whitegrid")
for column in numerical.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=numerical, x=column, kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()

# Using matplotlib for histograms
for column in numerical.columns:
    plt.figure(figsize=(6, 4))
    plt.hist(numerical[column], bins=30)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.show()
```

```python
# 3 - Check for multicollinearity between the numerical features.

# Calculate the correlation matrix
correlation_matrix = numerical.corr()

# Plot the correlation matrix using seaborn heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# High correlation pairs  with correlation greater than 0.9
high_correlation_pairs = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.9:
            col_i = correlation_matrix.columns[i]
            col_j = correlation_matrix.columns[j]
            high_correlation_pairs.add((col_i, col_j))

# Drop one feature from each high correlation pair
for feature_pair in high_correlation_pairs:
    print(f"Dropping one of the high correlation features: {feature_pair}")
    numerical.drop(feature_pair[0], axis=1, inplace=True)
```
